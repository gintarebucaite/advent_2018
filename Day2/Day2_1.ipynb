```python
with open('input', 'r') as day_2:
    d2 = day_2.readlines()
    
#print(d2)

def char_count(x):
    seen = {}
    for letter in x:
        if letter not in seen:
            seen[letter] = 1
        else:
            seen[letter] += 1
    print(seen)
    return seen

count_2 = 0
count_3 = 0

def check_count(x):
    global count_2
    global count_3
    temp_count_2 = 0
    temp_count_3 = 0
    letter_count = char_count(x)
    for letter in letter_count:
        if letter_count[letter] == 2:
            temp_count_2 += 1
        if letter_count[letter] == 3:
            temp_count_3 +=1
    if temp_count_2 > 0:
        count_2 +=1
    if temp_count_3 >0:
        count_3 +=1
    print(count_2)
    print(count_3)
        
    
for code in d2:
    check_count(code)
    
print(count_2)
print(count_3)

answer = count_2 * count_3
print(answer)
```
